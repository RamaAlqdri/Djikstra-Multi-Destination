# CDSSSD Notebook
**Algoritma:** Conventional Dijkstra Single Source Single Destination (CDSSSD)

## Tujuan
Menjalankan Dijkstra dari sumber yang sama ke setiap destination secara independen.

## Ringkasan Cara Kerja
- Setiap destination diproses satu per satu.
- Source tidak berubah.
- Destination duplikat tetap diproses berulang agar konsisten dengan urutan input.
- Jika `source` termasuk destination, hasilnya path satu node dengan cost 0.

```python
from pprint import pprint
from algorithms import dijkstra_shortest_path, run_cdsssd
```

```python
graph = {
    "A": [("B", 4), ("C", 2)],
    "B": [("A", 4), ("C", 1), ("D", 5), ("E", 12)],
    "C": [("A", 2), ("B", 1), ("D", 8), ("E", 10)],
    "D": [("B", 5), ("C", 8), ("E", 2), ("F", 6)],
    "E": [("B", 12), ("C", 10), ("D", 2), ("F", 2)],
    "F": [("D", 6), ("E", 2)],
}

source = "A"
destinations = ["C", "F", "E", "A", "C"]  # sengaja ada source + duplikat
```

```python
def print_separator(title: str) -> None:
    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)

print_separator("Helper + Dijkstra Demo")
demo = dijkstra_shortest_path(graph, source, "E")
pprint(demo)
```

```python
def run_algorithm(graph, source, destinations):
    """Wrapper implementasi CDSSSD dari module reusable."""
    return run_cdsssd(graph, source, destinations)
```

```python
print_separator("Demo Eksekusi CDSSSD")
result = run_algorithm(graph, source, destinations)
pprint(result)
```

```python
print_separator("Output Rapi CDSSSD")
print(f"Algorithm: {result['algorithm']}")
print(f"Source: {result['source']}")
print(f"Visit order: {result['visit_order']}")
print(f"Total cost: {result['total_cost']}")
print(f"Total visited nodes: {result['total_visited_nodes']}")
print("\nPer-destination result:")
for item in result['destination_results']:
    print(
        f"- to {item['destination']}: path={item['path']}, "
        f"cost={item['cost']}, visited={item['visited_nodes']}"
    )
```

## Penjelasan Hasil
CDSSSD menghitung shortest path dari `source` awal ke tiap destination tanpa
mewarisi posisi destination sebelumnya. Karena itu total cost adalah penjumlahan
dari beberapa query independen, bukan satu rute kontinu.
