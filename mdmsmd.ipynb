# MDMSMD Notebook
**Algoritma:** Modified Dijkstra Multi-Source Multi-Destination (MDMSMD)

## Tujuan
Menjalankan shortest path multi-destination berdasarkan urutan input destination.

## Ringkasan Cara Kerja
- Destination diproses sesuai urutan list.
- Setelah satu destination tercapai, node tersebut menjadi source baru.
- Sub-path digabung tanpa duplikasi node sambungan.
- Destination duplikat tetap dipertahankan sebagai permintaan terpisah.

```python
from pprint import pprint
from algorithms import dijkstra_shortest_path, run_mdmsmd
```

```python
graph = {
    "A": [("B", 4), ("C", 2)],
    "B": [("A", 4), ("C", 1), ("D", 5), ("E", 12)],
    "C": [("A", 2), ("B", 1), ("D", 8), ("E", 10)],
    "D": [("B", 5), ("C", 8), ("E", 2), ("F", 6)],
    "E": [("B", 12), ("C", 10), ("D", 2), ("F", 2)],
    "F": [("D", 6), ("E", 2)],
}

source = "A"
destinations = ["C", "F", "E", "A", "C"]  # sengaja ada source + duplikat
```

```python
def print_separator(title: str) -> None:
    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)

print_separator("Helper + Dijkstra Demo")
demo = dijkstra_shortest_path(graph, source, "F")
pprint(demo)
```

```python
def run_algorithm(graph, source, destinations):
    """Wrapper implementasi MDMSMD dari module reusable."""
    return run_mdmsmd(graph, source, destinations)
```

```python
print_separator("Demo Eksekusi MDMSMD")
result = run_algorithm(graph, source, destinations)
pprint(result)
```

```python
print_separator("Output Rapi MDMSMD")
print(f"Algorithm: {result['algorithm']}")
print(f"Full path: {result['full_path']}")
print(f"Visit order: {result['visit_order']}")
print(f"Total cost: {result['total_cost']}")
print(f"Total visited nodes: {result['total_visited_nodes']}")
print("\nSegments:")
for seg in result['segments']:
    print(
        f"- {seg['from']} -> {seg['to']}: path={seg['path']}, "
        f"cost={seg['cost']}, visited={seg['visited_nodes']}"
    )
```
