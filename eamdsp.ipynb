# EAMDSP Notebook
**Algoritma:** Efficient Algorithm for Multi-Destination Shortest Path (EAMDSP)

## Tujuan
Mencari urutan kunjungan destination secara greedy: setiap langkah memilih
destination terdekat dari source aktif.

## Ringkasan Cara Kerja
- Di tiap iterasi, hitung shortest path ke semua destination tersisa.
- Pilih destination dengan cost minimum.
- Destination terpilih menjadi source baru.
- Jika ada tie cost, dipilih yang muncul lebih dulu pada urutan destination tersisa.

```python
from pprint import pprint
from algorithms import dijkstra_shortest_path, run_eamdsp
```

```python
graph = {
    "A": [("B", 4), ("C", 2)],
    "B": [("A", 4), ("C", 1), ("D", 5), ("E", 12)],
    "C": [("A", 2), ("B", 1), ("D", 8), ("E", 10)],
    "D": [("B", 5), ("C", 8), ("E", 2), ("F", 6)],
    "E": [("B", 12), ("C", 10), ("D", 2), ("F", 2)],
    "F": [("D", 6), ("E", 2)],
}

source = "A"
destinations = ["C", "F", "E", "A", "C"]  # sengaja ada source + duplikat
```

```python
def print_separator(title: str) -> None:
    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)

print_separator("Helper + Dijkstra Demo")
demo = dijkstra_shortest_path(graph, source, "D")
pprint(demo)
```

```python
def run_algorithm(graph, source, destinations):
    """Wrapper implementasi EAMDSP dari module reusable."""
    return run_eamdsp(graph, source, destinations)
```

```python
print_separator("Demo Eksekusi EAMDSP")
result = run_algorithm(graph, source, destinations)
pprint(result)
```

```python
print_separator("Output Rapi EAMDSP")
print(f"Algorithm: {result['algorithm']}")
print(f"Full path: {result['full_path']}")
print(f"Visit order: {result['visit_order']}")
print(f"Total cost: {result['total_cost']}")
print(f"Total visited nodes: {result['total_visited_nodes']}")
print("\nSegments:")
for seg in result['segments']:
    print(
        f"- {seg['from']} -> {seg['to']}: path={seg['path']}, "
        f"cost={seg['cost']}, visited={seg['visited_nodes']}"
    )
```

## Penjelasan Hasil
EAMDSP tidak mengikuti urutan destination input secara kaku, tetapi
memilih tujuan terdekat dari posisi aktif di setiap iterasi.
Pendekatan ini biasanya menurunkan cost total dibanding urutan statis,
meski tidak selalu optimal global seperti TSP.
